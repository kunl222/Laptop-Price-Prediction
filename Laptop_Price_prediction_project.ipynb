## The next column is Memory column

```python
df['Memory'].value_counts()
```

```python
df['Memory'] = df['Memory'].astype(str).replace(r'\.0', '', regex=True)
df["Memory"] = df["Memory"].str.replace('GB', '')
df["Memory"] = df["Memory"].str.replace('TB', '000')
new = df["Memory"].str.split("+", n = 1, expand = True)

df["first"]= new[0]
df["first"]=df["first"].str.strip()

df["second"]= new[1]

df["Layer1HDD"] = df["first"].apply(lambda x: 1 if "HDD" in x else 0)
df["Layer1SSD"] = df["first"].apply(lambda x: 1 if "SSD" in x else 0)
df["Layer1Hybrid"] = df["first"].apply(lambda x: 1 if "Hybrid" in x else 0)
df["Layer1Flash_Storage"] = df["first"].apply(lambda x: 1 if "Flash Storage" in x else 0)

df['first'] = df['first'].str.replace(r'\D', '')

df["second"].fillna("0", inplace = True)

df["Layer2HDD"] = df["second"].apply(lambda x: 1 if "HDD" in x else 0)
df["Layer2SSD"] = df["second"].apply(lambda x: 1 if "SSD" in x else 0)
df["Layer2Hybrid"] = df["second"].apply(lambda x: 1 if "Hybrid" in x else 0)
df["Layer2Flash_Storage"] = df["second"].apply(lambda x: 1 if "Flash Storage" in x else 0)

df['second'] = df['second'].str.replace(r'\D', '')

# Extract digits only, assume size like '128 SSD', '1 TB HDD', etc.
df["first"] = df["first"].str.extract(r'(\d+)')  # this extracts only the number part
df["second"] = df["second"].str.extract(r'(\d+)')

# Fill NaNs and cast to int
df["first"] = df["first"].fillna(0).astype(int)
df["second"] = df["second"].fillna(0).astype(int)


df["HDD"]=(df["first"]*df["Layer1HDD"]+df["second"]*df["Layer2HDD"])
df["SSD"]=(df["first"]*df["Layer1SSD"]+df["second"]*df["Layer2SSD"])
df["Hybrid"]=(df["first"]*df["Layer1Hybrid"]+df["second"]*df["Layer2Hybrid"])
df["Flash_Storage"]=(df["first"]*df["Layer1Flash_Storage"]+df["second"]*df["Layer2Flash_Storage"])

df.drop(columns=['first', 'second', 'Layer1HDD', 'Layer1SSD', 'Layer1Hybrid',
       'Layer1Flash_Storage', 'Layer2HDD', 'Layer2SSD', 'Layer2Hybrid',
       'Layer2Flash_Storage'],inplace=True)
```

### We know that the resultant column (i.e 'Price') is right skewed 
# What is the meaning of skewed 
### In statistics, "skewed" describes a distribution that is not symmetrical—it has a longer tail on one side than the other. This indicates that the data values are not evenly distributed around the mean.
### Types of Skewness:
## Right-Skewed (Positive Skew)

### Tail is longer on the right.

### Most values are concentrated on the left.

### Example: Income distribution (many people earn less, few earn a lot).

## Left-Skewed (Negative Skew)

### Tail is longer on the left.

### Most values are concentrated on the right.

### Example: Age at retirement (most people retire around 60–65, few retire much earlier).



## Symmetrical (No Skew)

### Mean = Median = Mode.

### Bell-shaped curve (like normal distribution).

## 🧠  Why Skewness Matters:
### Skewness can distort the mean and affect statistical analyses.

### Some machine learning algorithms (like linear regression) assume normality.

### Often, skewed data is transformed (e.g., log transformation) to make it more normal.

### Linear Regression

```python
step1 = ColumnTransformer(transformers=[
    ('col_tnf',OneHotEncoder(sparse_output=False,drop='first'),[0,1,7,10,11])
],remainder='passthrough')

step2 = LinearRegression()

pipe = Pipeline([
    ('step1',step1),
    ('step2',step2)
])

pipe.fit(X_train,y_train)
y_pred = pipe.predict(X_test)

print('R2 score',r2_score(y_test,y_pred))
print('MAE',mean_absolute_error(y_test,y_pred))
```

### Ridge Regression

####  👉 the ridge regression advantage is to avoid overfitting.
####  👉 Our ultimate model is the one that coulld generalize patterns. i.e that works best on training and testing dateset . 
####  👉 Overfitting occurs when the trained model performs well on the training data and performs poorly on the testing data .



```python
step1 = ColumnTransformer(transformers=[
    ('col_tnf',OneHotEncoder(sparse_output=False,drop='first'),[0,1,7,10,11])
],remainder = 'passthrough')

step2 = Ridge(alpha = 1.4)

pipe = Pipeline([
    ('step1',step1),
    ('step2',step2)
])

pipe.fit(X_train,y_train)

y_pred = pipe.predict(X_test)

print('R2 score',r2_score(y_test,y_pred)*100)
print('MAE',mean_absolute_error(y_test,y_pred))
```

## What is the use of 'alpha' parameter ??

#### The alpha parameter in Ridge regression controls the amount of regularization applied to the model.

### Specifically:
#### Ridge regression = Linear Regression + L2 penalty
#### The L2 penalty = α × sum of squared coefficients (except the intercept)

### So:
#### If alpha = 0 → It's just regular linear regression (no regularization)
#### If alpha > 0 → It penalizes large coefficients, helping to reduce model complexity and prevent overfitting

### What does alpha = 10 mean?
#### You're applying moderate regularization.
#### The model will prefer smaller coefficients, which may generalize better to unseen data.
#### It can improve performance when there's multicollinearity or when you have more features than samples.

## Stacking Regressor

```python
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor, ExtraTreesRegressor, VotingRegressor
from sklearn.linear_model import Ridge
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
from sklearn.pipeline import Pipeline
from sklearn.metrics import r2_score, mean_absolute_error

# Define transformers
step1 = ColumnTransformer(transformers=[
    ('col_tnf', OneHotEncoder(sparse_output=False, drop='first'), [0, 1, 7, 10, 11])
], remainder='passthrough')

# Define models (note: XGBRegressor is NOT used)
rf = RandomForestRegressor(n_estimators=350, random_state=3, max_samples=0.5, max_features=0.75, max_depth=15)
gbdt = GradientBoostingRegressor(n_estimators=100, max_features=0.5)
et = ExtraTreesRegressor(n_estimators=100, random_state=3, max_features=0.75, max_depth=10)

# Define ensemble
step2 = VotingRegressor(estimators=[
    ('rf', rf),
    ('gbdt', gbdt),
    ('et', et)
], weights=[5, 1, 1])

# Create pipeline
pipe = Pipeline([
    ('step1', step1),
    ('step2', step2)
])

# Fit and evaluate
pipe.fit(X_train, y_train)
y_pred = pipe.predict(X_test)

print("R2 Score:", r2_score(y_test, y_pred))
print("MAE:", mean_absolute_error(y_test, y_pred))
```
